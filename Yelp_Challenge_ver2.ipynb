```python
df = pd.DataFrame(data_one)
```

```python
data_main = []
with open('F:\yelp\yelp_academic_dataset_user.json') as f:
    for line in f:
        data_main.append(json.loads(line))
```

```python
df_user = pd.DataFrame(data_main)
```

```python
merged_df = pd.merge(df, df_user, on='user_id', how='inner')
```

```python
merged_df.stars.value_counts(ascending=True)
```

```python
total_star_categories = 5
stars_labels = np.array([x_stars+1 for x_stars in range(total_star_categories)])
star_category_dist_fig = plt.figure(figsize=(12,8))
bar_plot_indices = np.arange(total_star_categories)
star_category_absolute_frequencies = merged_df.stars.value_counts(ascending=True)
star_category_relative_frequencies = np.array(star_category_absolute_frequencies)/float(sum(star_category_absolute_frequencies))

rects = plt.bar(bar_plot_indices, star_category_relative_frequencies, width=1, alpha=.7)

for (idx, rect) in enumerate(rects):
        plt.gca().text(rect.get_x()+rect.get_width()/2., 1.05*rect.get_height(), '%d'%int(star_category_absolute_frequencies[idx+1]),
                ha='center', va='bottom')
        
plt.xticks(bar_plot_indices+.5, stars_labels)
plt.xlabel('Star Category')
plt.ylabel('Relative Frequency')
plt.ylim([0,1])
plt.title('Star Category Distribution for {0} Reviews'.format(len(merged_df)))

plt.show()
```
